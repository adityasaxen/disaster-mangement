# 🤖 Model Training
## Training Pipeline for Cyclone, Drought, and Tornado

This notebook is part of the Disaster Prediction System MCA Project.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib
import os

# Define Centralized Data Path
DATA_PATH = r'C:\Users\adity\OneDrive\Desktop\disaster\data'

def train_save_model(file_name, target_col, model_name):
    # Construct full path
    full_path = os.path.join(DATA_PATH, file_name)
    df = pd.read_csv(full_path)
    
    # 1. Feature Engineering & Categorical Mapping
    if 'tornado_risk_category' in df.columns:
        # Mapping ordinal risk levels to numerical values for the classifier
        risk_map = {'Low': 0, 'Moderate': 1, 'High': 2, 'Very High': 3}
        df['target'] = df['tornado_risk_category'].map(risk_map)
        # Using Geospatial coordinates as primary predictors for Tornadoes
        X = df[['latitude', 'longitude']]
    else:
        df['target'] = df[target_col]
        # Automating feature selection by isolating numerical meteorological data
        X = df.select_dtypes(include=[np.number]).drop(['target'], axis=1, errors='ignore')
    
    # 2. Data Splitting (80% Train, 20% Test)
    X_train, X_test, y_train, y_test = train_test_split(
        X, df['target'], test_size=0.2, random_state=42
    )
    
    # 3. Model Implementation: Random Forest Classifier
    # Utilizing 100 decision trees to minimize variance and prevent overfitting
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    
    # 4. Persistence: Saving the trained brain for the Prediction Dashboard
    joblib.dump(model, model_name)
    print(f'Successfully saved: {model_name} (Input: {file_name})')

# Executing the Pipeline for all disaster vectors
train_save_model('cyclone.csv', 'cyclone_risk', 'cyclone_model.pkl')
train_save_model('drought.csv', 'drought_risk', 'drought_model.pkl')
train_save_model('tronado.csv', 'tornado_risk_category', 'tornado_model.pkl')
```
